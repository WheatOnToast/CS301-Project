```python
import os
import cv2
import cvlib as cv
import numpy as np
import matplotlib.pyplot as plt
from cvlib.object_detection import draw_bbox

```

```python
from ultralytics import YOLO
import cv2

# load yolov8 model
model = YOLO('yolov8n.pt')

# Display model information (optional)
model.info()

# Train the model on the COCO8 example dataset for 100 epochs
results = model.train(data='coco8.yaml', epochs=100, imgsz=640)

# Run inference with the YOLOv8n model on a drone image
results = model('drone.png')


#create detections
current_dir = os.getcwd()
det_dir_name = 'detections'
det_dir_path = os.path.join(current_dir, det_dir_name)

#create detections directory
if(not(os.path.exists(det_dir_path))): 
    os.makedirs(det_dir_path)

for video in os.listdir('videos'):
    indexed_frame = 0

    # load video
    cap = cv2.VideoCapture('videos/'+video) 

    #create detected frames folder for the video
    vid_det_path = os.path.join(det_dir_name, str(video))
    if(not(os.path.exists(vid_det_path))): 
        os.makedirs(vid_det_path)

    ret = True
    # read frames
    while ret:
        ret, frame = cap.read()

        if ret:
            # detect objects
            # track objects
            results = model.track(frame, persist=True)

            # plot results
            frame_ = results[0].plot()

            # write tracked images
            cv2.imwrite(os.path.join(vid_det_path, 'frame'+str(indexed_frame)+'.jpg'), frame_)

        indexed_frame += 1
 

```
